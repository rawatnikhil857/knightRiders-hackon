loading bloom-1b7 LM and our fine tuned model configuration from hugging face saved as "rawatnikhil857/movieRec-bloom-1b7".


```python
from peft import PeftModel, PeftConfig
from transformers import AutoTokenizer, AutoConfig, AutoModelForCausalLM
import pandas as pd
from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM

config = PeftConfig.from_pretrained("rawatnikhil857/movieRec-bloom-1b7")
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-1b7")
model = PeftModel.from_pretrained(model, "rawatnikhil857/movieRec-bloom-1b7")
```

```python
#loading tokenizer used for bloom LM
tokenizer = AutoTokenizer.from_pretrained("bigscience/tokenizer")
```

```python
import torch
```

```python
from IPython.display import display, Markdown

#function to generate results from given context(list of movies a user has already seen and their ratings) and question
def make_inference(context, question):
  batch = tokenizer(f"### CONTEXT\n{context}\n\n### QUESTION\n{question}\n\n### ANSWER\n",return_tensors='pt', return_token_type_ids=False)
  with torch.cuda.amp.autocast():
    output_tokens = model.generate(**batch, max_new_tokens=200, repetition_penalty=1.2)

  display(Markdown((tokenizer.decode(output_tokens[0], skip_special_tokens=False))))
```

# Predicted outputs

```python
data = pd.read_csv("Context-Question-AnswerDatabase.csv")
context = data["Context"][30]
question = "What will be some more movies this user might like?"
make_inference(context, question)
```
